```python
import cv2
import matplotlib.pyplot as plt
```

```python
puppy = cv2.imread("../notebooks/DATA/00-puppy.jpg")
plt.imshow(puppy)
```

```python
# Convert BGR -> RGB
puppy = cv2.cvtColor(puppy, code=cv2.COLOR_BGR2RGB)
plt.imshow(puppy)
```

```python
# Convert RGB -> HSV (Hue, Saturation, Value)
puppy = cv2.cvtColor(puppy, code=cv2.COLOR_RGB2HSV)
plt.imshow(puppy)
```

```python
# Convert HSV -> HLS (Hue, Light, Saturation)
puppy = cv2.cvtColor(puppy, code=cv2.COLOR_HSV2BGR)
puppy = cv2.cvtColor(puppy, code=cv2.COLOR_BGR2HLS)
plt.imshow(puppy)
```

```python
# Convert HLS -> RGB
puppy = cv2.cvtColor(puppy, code=cv2.COLOR_HLS2RGB)
plt.imshow(puppy)
```
