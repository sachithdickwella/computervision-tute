```python
import matplotlib.pyplot as plt

import cv2
```

```python
img = cv2.imread('../notebooks/DATA/00-puppy.jpg')
print(type(img))
print(img.shape)
```

```python
# If the file/image is not available in the given path, OpenCV doesn't throw an exception.
# Instead, it'll return 'NoneType' object.
img_wrong = cv2.imread('../../no/image.png')
type(img_wrong)
```

```python
# MATPLOTLIB expect color channel in the order of RGB
# OpenCV returns the color channel in the order of BGR.
plt.imshow(img)
# That's why here is the bluish discrepant color.
```

```python
# Simply swap the Channel dimension with the step to resolve that.
plt.imshow(img[:, :, ::-1])
```

```python
# OR just use the function OpenCV has provided. We can convert any color format to any color
# format using the 'cvtColor' function with different 'code'.
fix_img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# Read image directly as a Grayscale image from the storage. We can use different set of
# flags to read the image with different color profiles.
img = cv2.imread('../notebooks/DATA/00-puppy.jpg', flags=cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
plt.imshow(fix_img)
print(fix_img.shape)
```

```python
# RESIZE IMAGES ('dsize' is destination size which takes a tuple in order of
# (new_width, new_height)).
new_img = cv2.resize(fix_img, dsize=(1000, 400))
plt.imshow(new_img)
```

```python

# RESIZE IMAGE by RATIO.
w_ratio = 0.01 # Width ratio 3% of the original.
h_ratio = 0.01  # Height ratio 3% of the original.

new_img = cv2.resize(fix_img, dsize=(0, 0), dst=fix_img, fx=w_ratio, fy=h_ratio)
plt.imshow(new_img)
```

```python
# HORIZONTAL and VERTICAL FLIP.
# flipcode=0[Vertical Flip], flipcode=1[Horizontal Flip] and flipcode=-1(Both flips)
new_img = cv2.flip(fix_img, flipCode=0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, flipCode=1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, flipCode=-1)
plt.imshow(new_img)
```

```python
# SAVE the numpy array as an image.
print(type(fix_img))
is_saved: bool = cv2.imwrite('new_file.jpg', cv2.cvtColor(fix_img, code=cv2.COLOR_RGB2BGR))
# Convert the image from RGB -> BGR before save hence OpenCV's perspective, that's the
# correct color order.
is_saved
```
