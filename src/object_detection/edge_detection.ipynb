```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
sammy = cv2.imread('../../notebooks/DATA/sammy_face.jpg')
plt.imshow(sammy)
```

```python
'''
CANNY EDGE DETECTION.
'''
edge = cv2.Canny(image=sammy, threshold1=127, threshold2=127)
plt.imshow(edge)
```

```python
# Optimize the edge detection by adjusting threshold values.
edge = cv2.Canny(image=sammy, threshold1=0, threshold2=255)
plt.imshow(edge)
```

```python
# Calculate optimal values for thresholds.
median = np.median(sammy) # 64.0
# Get the max from either 0 or 70% of median.
lower_bound = int(max(0, median * 0.7))
# Get the minimum from either 255 or 130% of median.
uper_bound = int(min(255, median * 1.3))
```

```python
edge = cv2.Canny(image=sammy, threshold1=lower_bound, threshold2=uper_bound * 5)
plt.imshow(edge)
```

```python
# Try blurring the image.
blurred = cv2.blur(src=sammy, ksize=(5, 5))
plt.imshow(blurred)
```

```python
edge = cv2.Canny(image=blurred, threshold1=lower_bound, threshold2=uper_bound + 50)
plt.imshow(edge)
```
